```python
from pandas import DataFrame
from py2neo import Graph,Node,Relationship,NodeMatcher
import pandas as pd
import numpy as np
import os
graph=Graph('http://localhost:7474/db/data/',auth=("neo4j","123456"))
#graph.run('MATCH (n) OPTIONAL MATCH (n)-[r]-() DELETE n,r')

#读取数据
stock = pd.read_csv('D:\大二下\创新创业社会实践\股票基础数据.csv',encoding="gbk") #股票
holder = pd.read_csv('D:\大二下\创新创业社会实践\股东数据.csv',encoding="gbk") #股东
concept = pd.read_csv('D:\大二下\创新创业社会实践\股票概念数据.csv',encoding="gbk") #概念
news = pd.read_csv('D:\大二下\创新创业社会实践\股票公告数据.csv',encoding="gbk") #公告
fund = pd.read_csv('D:\大二下\创新创业社会实践\持有基金数据.csv',encoding="gbk") #基金
```

```python
#创建股票实体
for i in stock.values:
    a=Node('股票',股票代码=i[0],股票名称=i[1],行业=i[3],公司名称=i[2],所在地=i[4],法人代表=i[6],联系电话=i[7])
    print('股票代码:'+str(i[0]),'股票名称:'+str(i[1]),'行业:'+str(i[3]),'公司名称:'+str(i[2]),'所在地:'+str(i[4]),'法人代表:'+str(i[6]),'联系电话:'+str(i[7]))
    graph.create(a)
    
#创建股东实体
for i in holder.values:
    b=Node('股东',股票代码=i[0],股东名称=i[1],股东性质=i[2],股份类型=i[3],持股数=i[4],占总股本比例=i[5])
    print('股票代码:'+str(i[0]),'股东名称:'+str(i[1]),'股东性质:'+str(i[2]),'股份类型:'+str(i[3]),'持股数:'+str(i[4]),'占总股本比例:'+str(i[5]))
    graph.create(b)

#创建基金实体
for i in fund.values:
    c=Node('基金',股票代码=i[0],基金代码=i[1],基金名称=i[2],持有数量=i[3],持仓市值=i[4],占总股本比例=i[5])
    print('股票代码:'+str(i[0]),'基金代码:'+str(i[1]),'基金名称:'+str(i[2]),'持有数量:'+str(i[3]),'持仓市值:'+str(i[4]),'占总股本比例:'+str(i[5]))
    graph.create(c)
    
#创建概念实体
for i in concept.values:
    d=Node('概念',股票代码=i[0],概念=i[1])
    print('股票代码:'+str(i[0]),'概念:'+str(i[1]))
    graph.create(d)

#创建公告实体
for i in news.values:
    e=Node('公告',股票代码=i[0],公告发布日期=i[1],标题=i[2])
    print('股票代码:'+str(i[0]),'公告发布日期:'+str(i[1]),'标题:'+str(i[2]))
    graph.create(e)
    
```

```python
#创建关系：参股（股票，股东）
matcher = NodeMatcher(graph)
for i in holder.values:
    a = matcher.match("股票",股票代码=i[0]).first()
    b = matcher.match("股东",股票代码=i[0])
    for j in b:
        r = Relationship(j,'参股',a)
        graph.create(r)
        print('股票代码',str(i[0]))
```

```python
#创建关系：持股（股票，基金）
matcher = NodeMatcher(graph)
for i in fund.values:
    a = matcher.match("股票",股票代码=i[0]).first()
    c = matcher.match("基金",股票代码=i[0])
    for j in c:
        r = Relationship(j,'持股',a)
        graph.create(r)
        print('股票代码',str(i[0]))
```

```python
#创建关系：发布公告（股票，公告）
matcher = NodeMatcher(graph)
for i in news.values:
    a = matcher.match("股票",股票代码=i[0]).first()
    e = matcher.match("公告",股票代码=i[0])
    for j in e:
        r = Relationship(j,'发布公告',a)
        graph.create(r)
```

```python
#创建关系：概念属于（股票，概念）
matcher = NodeMatcher(graph)
for i in concept.values:
    a = matcher.match("股票",股票代码=i[0]).first()
    d = matcher.match("概念",股票代码=i[0])
    for j in d:
        r = Relationship(j,'概念属于',a)
        graph.create(r)
```
